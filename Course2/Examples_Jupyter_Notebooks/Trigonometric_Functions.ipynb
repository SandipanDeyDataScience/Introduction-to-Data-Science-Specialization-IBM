## Small plotting exercise

This notebook includes the code to plot some trigonometric functions and customizes it with the different features of matplotlib.

a) Plot function of $sin(x)$

$$
f(x) = sin(x)
$$

```python
%matplotlib inline
%config InlineBackend.figure_format = "svg"

import numpy as np                     #Import the tools 
import matplotlib.pyplot as plt 
pi = np.pi
x = np.linspace(-np.pi, np.pi, 1000)    #Define the values of the x-axis
plt.rcParams["figure.figsize"] = (9,3)  #Tuple to configure the size of the plott
plt.xticks([-pi, -pi/2, 0, pi/2, pi],   #Labels for the x-axis
            ["$-\pi$", "$-\pi/2$", "$0$", "$\pi/2$", "$\pi$"])
plt.axhline(0,color="black", lw=1)      #Origin line
plt.plot(x, np.sin(x), label=r"$f(x)=\sin(x)$")  #Plot the function
plt.title("The $sin(x)$ function")
plt.xlabel("Radians")
plt.legend(loc="best", fontsize=15)
plt.show()                                       
```

d) Plot of the function $sinc(x)$

The sinc function $sinc(x)$, also called the "sampling function," is a function that arises frequently in signal processing and the theory of Fourier transforms. The full name of the function is "sine cardinal," but it is commonly referred to by its abbreviation, "sinc." 

$$
f(x) =\frac{\sin(x)}{x}
$$

```python
%matplotlib inline
%config InlineBackend.figure_format = "svg"

import numpy as np                     #Import the tools
import matplotlib.pyplot as plt 
pi = np.pi
x = np.linspace(-6*np.pi, 6*np.pi, 1000)
plt.rcParams["figure.figsize"] = (9, 4)  #Tuple to configure the size of the plott
plt.axhline(0,color="black", lw=1)      #Origin line
plt.xticks([-6*pi, -4*pi, -2*pi, 0, 2*pi, 4*pi, 6*pi],
          ["$-6\pi$","$-4\pi$", "$-2\pi$", "$0$", "$2\pi$", "$4\pi$", "$6\pi$" ])
plt.plot(x, np.sin(x)/x, label=r"$f(x)=\sinc(x)$")
plt.title("The $sinc(x)$ function")
plt.xlabel("Radians")
plt.legend(loc="best", fontsize=15)
plt.show()
```
