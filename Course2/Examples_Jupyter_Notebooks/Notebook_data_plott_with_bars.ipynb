# Code to represent categorical data with rectangular bars

 The following code represents categorical data with rectangular bars. The information was obtained from real statistics from [Project Euler](https://projecteuler.net/)

```python
%matplotlib inline

import matplotlib.pyplot as plt 

# Our data
labels = ["C/C++", "Java", "Python", "Haskell", "#C"]
members = [42885, 29001 , 50505, 6791, 13517]

# Generating the y positions. Later, we'll use them to replace them with labels.
y_positions = range(len(labels))

# Creating our bar plot
plt.bar(y_positions, members)
plt.xticks(y_positions, labels)
plt.ylabel("Members")
plt.title("Most used Programming language to solve Project Euler")
plt.show()
```
